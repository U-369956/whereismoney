```python
from chinese_calendar import is_workday, is_holiday
from datetime import date
from datetime import timedelta
import datetime
    
    
    
def is_work_day():
    april_last = date(2020, 10, 10)   # datetime.date
    print(is_workday(april_last))     # True
    print(is_holiday(april_last))     # False
    print(april_last.weekday())       # 5-星期六
 
 
# 当前日期N天前的证券交易日
def get_trade_day(n ,date):
    dt = datetime.datetime.strptime( str(date),'%Y%m%d')
    trade_day = '20121015'
    if n < 0:
        t = -n
    else:
        t = n
    for i in range(100):
        if n<0:
            delta_day = timedelta(days=-i)
        else:
            delta_day = timedelta(days=i)
        trade_day = dt-delta_day
        if is_workday(trade_day) and trade_day.weekday()<5:       # 工作日并且不是周末
            if t ==0:
                break
            t = t -1
    print(trade_day.strftime('%Y%m%d')) , str(date)
    return trade_day.strftime('%Y%m%d') , str(date)
```
