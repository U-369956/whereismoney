```python
import pandas as pd
import numpy  as np
from chinese_calendar import is_workday, is_holiday
from datetime import date
from datetime import timedelta
import datetime
```

```python
def get_industry( industry_file):
    file = pd.read_csv(  industry_file  ,engine='python', header=0, sep=',',index_col=None)
    file = file.drop( ['生效日期','证券英文名称','证监会行业分类代码','证监会行业分类英文名称','证监会行业分类代码(全)','证监会行业分类名称(全)','证监会行业分类英文名称(全)'] , axis=1 )
    stock_id = []
    for i in file.index:
        if file.loc[i]['交易所']== 'Shenzhen':
            stock_id.append(  "%06d" %file.loc[i]['证券代码'] + '.sz' )
        else:
            stock_id.append(  "%06d" %file.loc[i]['证券代码'] + '.sh' )
    file['stock_id'] = stock_id
    list1 = file['证监会行业分类名称'].values.tolist()
    industry = list( set( list1   ) )
    industry.sort(key=list1.index)
    file = file.set_index("证监会行业分类名称", drop=True)
    return industry,file
    
```
